```python
search_key = "karen's diner"

csvfile = open(search_key+".csv", "a+", newline="", encoding="utf-8")
csvwriter = csv.writer(csvfile)
u = []
t = []

for tweet in tweepy.Cursor(api.search, q = search_key, count =1000, lang="id").items(1000):
  # print(tweet.user.name, tweet.text)
  u.append(tweet.user.name)
  t.append(tweet.text.encode("utf-8"))
  tweets = [tweet.user.name, tweet.text.encode("utf-8")]
  csvwriter.writerow(tweets)

dictTweets = {"username":u, "teks":t}
df = pd.DataFrame(dictTweets, columns = ["username", "teks"])

df
```

```python
df = df.drop_duplicates(subset= 'teks', keep='first', inplace=False)
df
```

```python
def cleanTxt(text):
  text = re.sub(r'@[A-Za-z0-9]+','',text) #Remove @mentions
  text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
  text = re.sub(r'#', '', text) #Removing the '#' simbol
  text = re.sub(r'RT[\s]+', '', text) #Removing RT
  text = re.sub(r'https?:\/\/S+', '', text) #Remove the hyperlink
  text = re.sub(r'[^\w]|_',' ',text) #remove punctuations
  text = re.sub(r'b[\s]+', '', text) #Removing RT

  return text

#Cleaning the text

df['teks'] = df['teks'].apply(cleanTxt)

#Show the cleaned text
df
```
